```python
import sys

def welcome_page():
    print("""
    Welcome to your checkbook!
    """
    )

def home_page():
    print(''' 
    Transactions
    1 - View current balance
    2 - Record a credit (deposit)
    3 - Record a debit (withdraw)
    4 - Exit
    ''')

    user_input = input('Enter your desired transaction: ')
    user_input = int(user_input)
    
    if user_input == 1:
        view_balance()
    elif user_input == 2:
        deposit()
    elif user_input == 3:
        withdrawal()
    elif user_input == 4:
        view_history()
    elif user_input == 5:
        exit_program()
    else:
        print('Sorry, we could not process your selection. If you wish to continue, please select the number of your desired transaction.')
        print()
        home_page()
        
### Functions to normalize signs

def make_positive(n):
    return abs(float(n))
    
def make_negative(n):
    n *= -1.0
    return n
    
### Transaction Functions

def view_balance():

    vb_credit = 0
    vb_debit = 0

    with open('user_transactions.txt') as ut:
        t_reader = ut.read().split('\n')
        
        for item in t_reader:
            if item.startswith('+'):
                vb_credit += float(item.strip('+'))
            elif item.startswith('-'):
                vb_debit += float(item.strip('-'))

    current_balance = vb_credit - vb_debit
    
    print(f"Your current balance is: ${current_balance}")
    home_page()

def deposit():
    deposit_amount = input("How much would you like to deposit? $")
    
    # Condition: digits only
    if deposit_amount.isdigit() == True:
        deposit_amount = make_positive(deposit_amount)
        with open('user_transactions', 'a+') as ut:
            ut.write(str(deposit_amount) + '\n')
            print("You've successfully made a deposit. Thank you for your business.")
    else:
        print("Incorrect value was received. Please try again.")
    
    ut.close()
    home_page()
    
    # Main calculation
    


def withdraw():
    withdrawal_amount = input("How much would you like to withdraw? $")
    
    # Condition: digits only
    if withdrawal_amount.isdigit() == True:
        withdrawal_amount = withdrawal_amount.replace(",","")
        withdrawal_amount = make_negative(withdrawal_amount)
    else:
        print("Incorrect value was received. Please try again.")
    home_page()
    
    # Main calculation
    
    with open('user_transactions', 'a+') as ut:
        ut.write(withdrawal_amount + '\n')
    
    print('Withdrawal has successfully been made. Thank you for your business.')
    home_page()


def exit_program():
    exit_choice = input ("Are you want to exit? Y/N: ")
    exit_choice = exit_choice.lower()
    
    # Condition: normalization
    if exit_choice == "y":
        sys.exit()
    elif exit_choice == "n":
        display_menu()
    else:
        print("Invalid input")
        home_page()

welcome_page()
home_page()
```
